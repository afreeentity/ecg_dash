Synthetic ECG Dataset Generator for Arrhythmia Detection

This script generates realistic synthetic ECG data with labeled arrhythmias
for training and testing the deep learning model.

Generates 6 types of ECG rhythms:
1. Normal Sinus Rhythm 
2. Atrial Fibrillation
3. Bradycardia
4. Tachycardia
5. Premature Ventricular Contractions (PVCs)
6. Irregular Rhythm
"""

```python
class ECGGenerator:
    """Generate realistic synthetic ECG signals"""
    
    def __init__(self, sampling_rate=250):
        """
        Initialize ECG generator
        
        Args:
            sampling_rate (int): Samples per second (Hz)
        """
        self.fs = sampling_rate
        
    def generate_qrs_complex(self, duration=0.08):
        """
        Generate a QRS complex (the main spike in ECG)
        
        Args:
            duration (float): Duration of QRS in seconds
            
        Returns:
            np.array: QRS complex waveform
        """
        n_samples = int(duration * self.fs)
        t = np.linspace(-duration/2, duration/2, n_samples)
        
        # Q wave (small negative deflection)
        q_wave = -0.2 * np.exp(-((t + duration/4) ** 2) / (2 * (duration/10) ** 2))
        
        # R wave (large positive spike)
        r_wave = 1.0 * np.exp(-(t ** 2) / (2 * (duration/8) ** 2))
        
        # S wave (small negative deflection)
        s_wave = -0.15 * np.exp(-((t - duration/4) ** 2) / (2 * (duration/10) ** 2))
        
        qrs = q_wave + r_wave + s_wave
        return qrs
    def generate_p_wave(self, duration=0.08):
        """
        Generate a P wave (atrial depolarization)
        
        Args:
            duration (float): Duration in seconds
            
        Returns:
            np.array: P wave waveform
        """
        n_samples = int(duration * self.fs)
        t = np.linspace(-duration/2, duration/2, n_samples)
        p_wave = 0.15 * np.exp(-(t ** 2) / (2 * (duration/3) ** 2))
        return p_wave
    def generate_t_wave(self, duration=0.16):
        """
        Generate a T wave (ventricular repolarization)
        
        Args:
            duration (float): Duration in seconds
            
        Returns:
            np.array: T wave waveform
        """
        n_samples = int(duration * self.fs)
        t = np.linspace(-duration/2, duration/2, n_samples)
        t_wave = 0.2 * np.exp(-(t ** 2) / (2 * (duration/2.5) ** 2))
        return t_wave
    
    def add_baseline_wander(self, signal_length, amplitude= 0.05, freq= 0.3):
        """
        Add realistic baseline wander in a low-frequency drift mode.

        Args:
            signal_length (int): Number of samples
            ampliitude (float): Amplitude of wander
            freq (float): Frequency of wander in Hz

        Returns:
            np.array: Baseline wander signal
        """
        t = np.arrange(signal_length) / self.fs
        wander = amplitude * np.sin(2*np.pi*freq*t)
        wander += amplitude*0.5*np.sin(2*np.pi*freq*2.3*t)
        return wander
        
    
```
